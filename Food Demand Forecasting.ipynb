```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train=pd.read_csv('train.csv')
df_test=pd.read_csv('test.csv')
df_fulfilment=pd.read_csv('fulfilment_center_info.csv')
df_meal=pd.read_csv('meal_info.csv')
```

```python
df_train_with_ctr=pd.merge(df_train,df_fulfilment,on='center_id')
df_test_with_ctr=pd.merge(df_test,df_fulfilment,on='center_id')
```

```python
df_train_final=pd.merge(df_train_with_ctr,df_meal,on='meal_id')
df_test_final=pd.merge(df_test_with_ctr,df_meal,on='meal_id')
```

```python
df_train_final
```

```python
df_test_final
```

```python
df_train_final.drop(columns=['center_id','meal_id'],inplace=True)
df_test_final.drop(columns=['center_id','meal_id'],inplace=True)
```

```python
df_train_final.head()
```

```python
df_train_final.dtypes
```

```python
train_final_dummies=pd.get_dummies(df_train_final)
test_final_dummies=pd.get_dummies(df_test_final)

train_final_dummies.head()
```

```python
X=train_final_dummies.drop(columns=['num_orders'])
y=train_final_dummies['num_orders']

X_test=test_final_dummies
```

```python
X
```

```python
sns.heatmap(X.corr())
```

```python
Pred_id=X_test.id
columns_to_drop=['base_price','id']
X.drop(columns_to_drop,axis=1,inplace=True)
X_test.drop(columns_to_drop,axis=1,inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
mm=RandomForestRegressor(n_jobs=-1,n_estimators=200,oob_score=True)
```

```python
mm.fit(X.values,np.array(y))
```

```python
y_preds=mm.predict(X_test.values)
np.sum(y_preds<0)
```

```python
submission=pd.read_csv("sample_submission.csv")
submission.head()
submission['id']=Pred_id
submission['num_orders']=y_preds
pd.DataFrame(submission, columns=['id','num_orders']).to_csv('RandomForests.csv',index=False)
```
